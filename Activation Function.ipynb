```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x=np.arange(-10,10)
```

#Identity

```python
y=x
```

```python
plt.title("Identity")
plt.plot(y)
```

#Sigmoid

```python
y=1/(1+pow(np.e,-x))
```

```python
plt.title("Sigmoid")
plt.plot(y)
```

#Threshold

```python
y=[]
```

```python
for _ in range(len(x)):
    if x[_]>=0:
        y.append(1)
    else:
        y.append(0)
```

```python
plt.title("Threshold")
plt.plot(y)
```

#Tanh

```python
y=(2/(1+pow(np.e,-2*x)-1))
```

```python
plt.title("Tanh")
plt.plot(y)
```

#ArcTan

```python
y=np.arctan(x)
plt.title("ArcTan")
plt.plot(y)
```

#ReLU

```python
y=[]
for _ in x:
    if _>=0:
        y.append(_)
    else:
        y.append(0)
```

```python
plt.title("ReLU")
plt.plot(y)
```

```python
#leaky ReLU
y=[]
for _ in x:
    if _>=0:
        y.append(_)
    else:
        y.append(0.1*_)
```

```python
plt.title("Leaky ReLU")
plt.plot(y)
```
